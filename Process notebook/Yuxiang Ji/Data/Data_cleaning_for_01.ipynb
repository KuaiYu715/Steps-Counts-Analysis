```python
p1_data.insert(0,'week','')
```

```python
p1_data['Date'].replace('-','',regex=True,inplace=True)
p1_data['Date'] = pd.to_datetime(p1_data['Date'], format='%Y%m%d')
p1_data['week'] = p1_data['Date'].dt.isocalendar().week # Calculate the wrrk number by python method.
p1_data
```

```python
p1_data.groupby('week')['Count'].sum().to_csv("/content/drive/MyDrive/Colab Notebooks/weeks_and_steps_1.csv",index=False,sep=',')
```

```python
p1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/weeks_and_steps_1.csv",sep = ',')
p1.head()
```

```python
p1.insert(0,'week','')
p1.insert(2,'sedentary','')
```

```python
for n in range(0,len(p1)):
  p1['week'][n] = n+1
p1.head()
```

```python
p1['week'] = p1['week'].astype('int')
p1['Count'] = p1['Count'].astype('int')
```

```python
for n in range(0,len(p1)):
  if p1['Count'][n] >= 10000*7:
    p1['sedentary'][n] = 0
  else:
    p1['sedentary'][n] = 1
p1
```
