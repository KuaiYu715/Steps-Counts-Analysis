# Data Cleaning

Fill the missing value with 0 and sort the date in a ascending order.

```python
copy_data['Count'].fillna(0,inplace=True)
```

```python
copy_data.sort_values(by=['Date'], inplace=True)
```

```python
copy_data
```

Remove outliers and delete the column 'source' which we do not want

```python
copy_data=copy_data.iloc[3:]
copy_data=copy_data.loc[:,['Date','Hour','Count']]
copy_data
```

Convert the column 'Date' into a different form such that we can modify

```python
copy_data['Date'].replace('-','',regex=True,inplace=True)
copy_data['Date'] = pd.to_datetime(copy_data['Date'], format='%Y%m%d')
print(copy_data.dtypes)
```

Convert date into week in order to gain more insights by the week

```python
copy_data['Date'] = copy_data['Date'].dt.isocalendar().week
copy_data['Date'] = copy_data['Date'].astype(int)
for i in range(len(copy_data)):
  if copy_data['Date'].values[i]==52:
    copy_data['Date'].values[i]=0
```

Drop the data from last year

```python
copy_data['Date'] = copy_data['Date'].astype(str)
for i in range(len(copy_data)):
  if copy_data['Date'].values[i]=='0':
    copy_data['Date'].values[i]= 'NaN'
copy_data=copy_data.drop(copy_data[copy_data.Date=='NaN'].index)
```

```python
copy_data
```

Groupby the target attribute

```python
combined_data=copy_data.groupby('Date')['Date','Count'].sum()
```
