```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv(r"C:\Users\HP1000\Documents\Data3406\P05.csv")
data
```

```python
ls = []
for value in data['Steps (count)']:
    if value >= 50000:
        ls.append('Low risk')
    else:
        ls.append('High risk')
dic = {'dia':ls}
data['diabetes'] = pd.DataFrame(dic)
data
```

```python
data['nth week'] = data.index + 1
```

```python
x = data['nth week']
y = data['Steps (count)']

plt.plot(x,y)
plt.axhline(50000,label = 'risk',linestyle = ':',color = 'r')
plt.title('Walking steps for each Week')
plt.xlabel('Week')
plt.ylabel('Steps count')
plt.legend()

plt.show()
```
