# **Would people choose to walk based on season?**

What I want to explore is that the number of steps per week is affected by seasons. The following code is I divide the time into weeks and then define them as four seasons. Use the histogram chart for data visualization and compare the trend of weekly steps in the four seasons.

-------
**Participant 01** 

````python
df = pd.read_csv('/content/drive/MyDrive/Participant-01.csv') #read csv
df["Start"] = df["Start"].apply(pd.to_datetime)
df["Finish"] = df["Finish"].apply(pd.to_datetime)
df.iloc[:,2] = df.iloc[:,2].astype(int) # let all the numbers to be interger
weaksum = pd.DataFrame(df.resample('W', on='Start')['Steps (count)'].sum()).reset_index() # sum of the steps
gb_month = weaksum.groupby([weaksum['Start'].dt.isocalendar().week, weaksum['Start'].dt.month]).mean().rename_axis(['week', 'month']).reset_index() # group by the weeks on Start time
season = gb_month.copy()
# make some definitions of season
season['month'] = season['month'].replace([9,10,11], 'Spring')
season['month'] = season['month'].replace([12,1,2], 'Summer')
season['month'] = season['month'].replace([3,4,5], 'Fall')
season['month'] = season['month'].replace([6,7,8], 'Winter')
season.rename({"Steps (count)":"sum of steps"}, axis = 1, inplace = True) #rename the column of Steps (count)
season.rename({"month":"season"}, axis = 1, inplace = True)
gb_season = season.groupby(['season','week']).mean() # group by the season and to be the mean value
spring = gb_season.iloc[gb_season.index.get_level_values('season') == 'Spring'].reset_index() 
summer = gb_season.iloc[gb_season.index.get_level_values('season') == 'Summer'].reset_index()
fall = gb_season.iloc[gb_season.index.get_level_values('season') == 'Fall'].reset_index()
winter = gb_season.iloc[gb_season.index.get_level_values('season') == 'Winter'].reset_index()
fig, ax = plt.subplots(2, 2, figsize=(20, 15)) #set a 2x2 figure table to show the 4 different seasons

plt.subplot(2,2,1) # to be the first 1 (top left)
plt.hist(x = spring['week'], weights = spring['sum of steps'], bins = 25)
ax = plt.gca()
ax.set_ylim([0, 250000])
plt.title('SPRING')

plt.subplot(2,2,2) # to be the first 1 (top right)
plt.hist(x = summer['week'], weights = summer['sum of steps'], bins = 25)
ax = plt.gca()
ax.set_ylim([0, 250000])
plt.title('SUMMER')

plt.subplot(2,2,3) # to be the first 1 (bottom left)
plt.hist(x = fall['week'], weights = fall['sum of steps'], bins = 25)
ax = plt.gca()
ax.set_ylim([0, 250000])
plt.title('FALL')

plt.subplot(2,2,4) # to be the first 1 (bottom right)
plt.hist(x = winter['week'], weights = winter['sum of steps'], bins = 25)
ax = plt.gca()
ax.set_ylim([0, 250000])
plt.title('WINTER')

plt.suptitle('Average Steps during the weeks in Seasons', fontsize=30)
````

From the above figure, we can see that the distribution of steps in spring and summer is relatively uniform and similar, while in autumn and winter, there is a huge gap between the first few weeks of winter and other weeks, and only one week of data in autumn exists. I think This is related to the incompleteness of the data. The step count data after September is missing. We will continue to explore and solve this problem.

Because the data is not complete in one year, the data in Spring and Summer is missing and it is impossible to observe the trend, so I decided to delete the images them to compare the trend of weekly steps in Fall and Winter.

# **Does temperature affect people's willingness to walk?**

--------
**Particpant 05**

Because we want to explore the weekly step dataset. So we decided to draw a line graph that shows the trend from basic time series model: **step count(t) = Level(t) + Trend(t) + Seasonality(t) + Noise(t), where t = any day.** 

```
 Aggregate by weekly data 
```



````python
M5_clean=M5 
M5_clean=M5_clean.resample('W-Mon', on='Date').sum() # resample into week
M5_clean.head() # check the weekly step counts.
````



```
We plot the trend component from the time series model to discover how temperature affect people's willingness to walk.
```



````python
# Basic time series model: step count(t) = Level(t) + Trend(t) + Seasonality(t) + Noise(t), where t = any day
m4=M5_clean 
m4['date']=m4.index # Create the new column called date with index values. 
import statsmodels.api as sm
decomposition = sm.tsa.seasonal_decompose(m4.drop(columns=['date']), model = 'additive',period=4) # Using additive model with paramater 4 that stands for 4 weeks per month.
# plot trend on matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
fig, ax = plt.subplots() # Draw the graph
ax.grid(True)
plt.plot(m4.index, m4['Count'], c='blue', label="Original version") # The original trend in blue.
plt.plot(decomposition.trend.index, decomposition.trend, c='red', label="Soomth version") # The smoothed trend in red
plt.xlabel("Weeks")
plt.ylabel("Steps")
plt.legend() # Show the label
plt.title("The true trend vs smooth trend of step data")
plt.xticks(rotation=45)
````

* From trend graph above, we could see there is an incaresing smooth version of trend from week to week until May and July respectively. People would take less step counts in summer with high temperature and more steps with low temperature. 

We conclude that the trend of step count will start with an increasing tendency in date with lower temperature.According to the visualization above indicates March and April are autumn in Australia which would be a better preference for walking. Hence, peope would take less step counts in summer like January and more steps in autumn or winter, such as June.
